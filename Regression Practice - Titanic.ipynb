```python
lreg1 = logit("survived ~ Age + Pclass + Female + SibSp", titanic).fit()
print (lreg1.summary())
```

```python
params = lreg1.params
conf=lreg1.conf_int()
conf['OR'] = params
conf.columns = ['Lower CI', 'Upper CI', 'Odds Ratios']
print(np.exp(conf))
```
